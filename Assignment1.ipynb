# Feature Extraction

```python
import os, re
from gensim.parsing.preprocessing import remove_stopwords, strip_punctuation
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
def preprocess(texts):
    # tokenization
    texts = [re.findall(r'\w+', line.lower()) for line in texts]
    # remove stopwords
    texts = [remove_stopwords(' '.join(line)).split() for line in texts]
    # remove punctuation
    texts = [strip_punctuation(' '.join(line)).split() for line in texts]
    # remove words that are only one or two characters
    texts = [[token for token in line if len(token) >2] for line in texts]
    # remove numbers
    texts = [[token for token in line if not token.isnumeric()] for line in texts]
    # lemmatization
    lemmatizer = WordNetLemmatizer()
    texts = [[word for word in lemmatizer.lemmatize(' '.join(line), pos='v').split()] for line in texts]
                                                    
    return texts       
```

```python
#preprocessed_texts = preprocess(df_fake.Headline)
preprocessed_texts = preprocess(df.complaint_what_happened)
```

```python
from gensim.models.phrases import Phrases, Phraser
```

```python
phrases = Phrases(preprocessed_texts, min_count=1, threshold=0.8, scoring='npmi')
```

```python
bigram = Phraser(phrases)
```

```python
preprocessed_texts = list(bigram[preprocessed_texts])
```

```python
preprocessed_texts
```

```python
from gensim.corpora import Dictionary
```

```python
dictionary = Dictionary(preprocessed_texts)
```

```python
dictionary.filter_extremes(no_below=10, no_above=0.6)
```

```python
corpus = [dictionary.doc2bow(text) for text in preprocessed_texts]
```

```python
corpus
```

# Topic Modelling using LDA

```python
from gensim.models import LdaModel
```

```python
import time
start_time = time.time()
```

```python
model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=50, chunksize=1500, iterations=200, alpha='auto')
```

```python
end_time = time.time()
print("training time in seconnds= ", (end_time - start_time))
```

```python
model.save('models/lda_model.model')
```
